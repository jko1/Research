```python
import BeautifulSoup
import urllib2, os
from BeautifulSoup import BeautifulSoup as Soup
from datetime import datetime
import re

html = urllib2.urlopen('http://jmlr.org/proceedings/').read()
page = Soup(html)
volume = page.findAll('ul')[1].findAll('li')
cycle_28 = 0
cycle_32 = 0

suffix = ['th','nd', 'st', 'rd']
for vol in volume:
  vol_str = vol.renderContents()
  end = vol_str.find("<")-1
  date_str = vol_str[6:end].replace("of ", "") #one had "of"
  for elem in suffix: #Which suffix for number
    if elem in date_str[:5]: #check first chars, since overlap with month
      date= datetime.strptime( date_str, "%d"+elem+" %B %Y")
      format_date = date.strftime("%m%d%Y")
  link = vol.findAll('a')[0].get('href')
  vol_num = vol.findAll('a')[0].renderContents().replace("Volume ","")
  if vol_num=='28':
    cycle_28+=1
  if vol_num=='32':
    cycle_32+=1
  html = urllib2.urlopen('http://jmlr.org/proceedings/'+ link).read()
  page = Soup(html)
  print "On " + vol_num + "_" + format_date
  if not os.path.exists( "/home/snifloo/jmlr/v"+ vol_num+ "_" + format_date):
    os.makedirs("/home/snifloo/jmlr/v"+ vol_num+ "_" + format_date)
  if int(vol_num)>=28:
    if vol_num=='28':
        page = page.findAll('dl')[cycle_28-1]
    if vol_num=='32':
        page = page.findAll('dl')[cycle_32-1]
    papers = page.findAll(attrs={'class':'paper'})
    for item in papers:
      title = item.findAll(attrs={'class':'title'})[0].renderContents().replace(" ", "-").replace("<span>","").replace("</span>","")
      title = re.sub(r'[^a-zA-Z0-9\[\]\(\)\s\-\_]','', title)
      pdfs = item.findAll(attrs = {'class':'links'})[0].findAll('a')[1::]
      for pdf in pdfs:
        pdf_link = pdf.get('href')
        if "/" not in pdf_link:
          pdf_link = 'http://jmlr.org/proceedings/papers/v' + vol_num + "/"+pdf_link  
      temp_pdf = re.findall('[^/]+$', pdf_link)[0].replace('.pdf','')
      file_name = "v"+vol_num + "-" +format_date + "-"+ temp_pdf + "_" + title 
      print pdf_link
      u = urllib2.urlopen(pdf_link)
      localFile = open("/home/snifloo/jmlr/v" + vol_num+ "_" + format_date +"/" + file_name + ".pdf", 'w')
      localFile.write(u.read())
      localFile.close()
  else:
    papers = page.findAll('dl')
    for item in papers:
      title = item.findAll('dt')[0].renderContents().strip()
      if "</font>" in title:
        index = title.find("<font")
        title = title[:index]
      title = re.sub(r'[^a-zA-Z0-9\[\]\(\)\s\-\_]','', title).replace(" ", "-")
      hrefs= item.findAll('a')
      for link in hrefs:
        if "pdf" in link.renderContents():
          pdf_link = link.get('href')
      if pdf_link[0:2]=="..":
        pdf_link= "http://jmlr.org/proceedings/papers" + pdf_link[2::]
      if pdf_link[0]==".":
        pdf_link= "http://jmlr.org/proceedings/papers/v" +vol_num+ "/"+ pdf_link[1::]
      if "http" not in pdf_link:
        pdf_link = "http://jmlr.org/proceedings/papers/v" + vol_num + "/"+pdf_link
      temp_pdf = re.findall('[^/]+$', pdf_link)[0].replace('.pdf','')
      file_name = "v" + vol_num + "-" + format_date + "-" + temp_pdf +"_"+ title
      print pdf_link
      u = urllib2.urlopen(pdf_link)
      localFile = open("/home/snifloo/jmlr/v" + vol_num+ "_" + format_date +"/" + file_name + ".pdf", 'w')
      localFile.write(u.read())
      localFile.close()
```
